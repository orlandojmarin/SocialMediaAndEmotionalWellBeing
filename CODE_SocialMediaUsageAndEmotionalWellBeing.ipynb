```python
# ORLANDO

import pandas as pd
import matplotlib.pyplot as plt

# save the training data path under the variable training_data_url
training_data_url = "https://raw.githubusercontent.com/orlandojmarin/SocialMediaAndEmotionalWellBeing/refs/heads/main/train.csv"

# read the csv file
df = pd.read_csv(training_data_url)

# show the first few rows of data
df.head()
```

```python
# show how many people are using each platform
df["Platform"].value_counts()
```

```python
# ORLANDO
# identify how much data is missing from each column in the dataset
null1 = df.isnull().sum()
print(null1)

# remove rows from the dataframe that have missing values
df = df.dropna()

```

```python
# ORLANDO
# identify how much data is missing from each column in the dataset
null2 = df.isnull().sum()
print(null2)

# print the number of rows with missing values
print(f"Number of rows with missing values: {df.isnull().any(axis=1).sum()}")
```

```python
# ORLANDO
# list every value under the age column of the dataset
df["Age"].unique()

# identify which values under the age column of the dataset are not numbers
non_numeric_age = df[df["Age"].str.isnumeric() == False]

# list every value under the gender column of the dataset
unique_genders = df["Gender"].unique()
print("Unique genders: ")
print(unique_genders)
print()

# identify which values under the gender column of the dataset are numeric
numeric_gender = df[df["Gender"].str.isnumeric() == True]

# print the number of rows in non_numeric_age dataframe
print(f"Number of rows in non_numeric_age: {len(non_numeric_age)}")

# print the number of rows in the numeric_gender dataframe
print(f"Number of rows in numeric_gender: {len(numeric_gender)}")

# list every value under the platform column of the dataset
unique_platforms = df["Platform"].unique()
print("\nUnique platforms: ")
print(unique_platforms)

# list every value under the Daily_Usage_Time (minutes) column
unique_daily_usage = df["Daily_Usage_Time (minutes)"].unique()
print("\nUnique daily usage: ")
print(unique_daily_usage)

# list every value under the Posts_Per_Day column
unique_posts_per_day = df["Posts_Per_Day"].unique()
print("\nUnique posts per day: ")
print(unique_posts_per_day)

# list every value under the Likes_Received_Per_Day column
unique_likes_received = df["Likes_Received_Per_Day"].unique()
print("\nUnique likes received: ")
print(unique_likes_received)

# list every value under the Comments_Received_Per_Day column
unique_comments_received = df["Comments_Received_Per_Day"].unique()
print("\nUnique comments received: ")
print(unique_comments_received)

# list every value under the Messages_Sent_Per_Day column
unique_messages_sent = df["Messages_Sent_Per_Day"].unique()
print("\nUnique messages sent: ")
print(unique_messages_sent)

# list every value under the Dominant_Emotion column
unique_emotions = df["Dominant_Emotion"].unique()
print("\nUnique emotions: ")
print(unique_emotions)
```

```python
# ORLANDO
# DATA CLEANING

# remove duplicate rows from the dataframe
df = df.drop_duplicates()

# calculate and print the number of duplicate rows from the dataframe
duplicated_rows = df.duplicated().sum()
print(f"Number of duplicated rows: {duplicated_rows}")

# Check if the non_numeric_age and numeric_gender dataframes
# have the same number of rows
print(len(non_numeric_age), len(numeric_gender))

# Ensure both datasets have the same length before proceeding
if len(non_numeric_age) == len(numeric_gender):
    # Swap 'Age' from non_numeric_age with 'Gender' from numeric_gender
    df.loc[non_numeric_age.index, 'Age'] = numeric_gender['Gender'].values
    df.loc[numeric_gender.index, 'Gender'] = non_numeric_age['Age'].values
else:
    print("Mismatch in the number of rows between datasets.")

# Step 5: Verify the update
print(df.head())

# show the unique values in the age column
new_unique_age = df["Age"].unique()
print("Unique Age: ")
print(new_unique_age)

print()

# show the unnique values in the gender column
new_unique_gender = df["Gender"].unique()
print("Unique Gender: ")
print(new_unique_gender)

# create a new column called "Activity_Score" by taking the sum of the "Posts_Per_Day", "Likes_Received_Per_Day", "Comments_Received_Per_Day", and "Messages_Sent_Per_Day" columns
df["Activity_Score"] = df["Posts_Per_Day"] + df["Likes_Received_Per_Day"] + df["Comments_Received_Per_Day"] + df["Messages_Sent_Per_Day"]

# print the first 5 rows of the updated dataframe
df.head()



```

```python
# ORLANDO
# convert the age column to a numeric datatype
df['Age'] = pd.to_numeric(df['Age'])

# show the datatypes of each column in the dataframe
df.dtypes
```

```python
# TATIANA
# ONE-HOT ENCODING

# one-hot encode the Gender column where the newly generated column names
# indicate if that column refers to Male, Female, or Non-binary and
# have the values in the columns be 1 for true or 0 for false
df_gender_one_hot_encoded = pd.get_dummies(df, columns=['Gender'], drop_first=False)

# convert True/False to 1/0
df_gender_one_hot_encoded[['Gender_Female', 'Gender_Male', 'Gender_Non-binary']] = df_gender_one_hot_encoded[['Gender_Female', 'Gender_Male', 'Gender_Non-binary']].replace({True: 1, False: 0})

# display the resulting DataFrame
print(df_gender_one_hot_encoded.head())
```

```python
# TATIANA
# ONE-HOT ENCODING

# one-hot encode the Platform column
df_platform_one_hot_encoded = pd.get_dummies(df_gender_one_hot_encoded, columns=['Platform'], drop_first=False)

# convert True/False to 1/0
df_platform_one_hot_encoded[['Platform_Facebook', 'Platform_Instagram', 'Platform_LinkedIn', 'Platform_Snapchat', 'Platform_Telegram', 'Platform_Twitter', 'Platform_Whatsapp']] = df_platform_one_hot_encoded[['Platform_Facebook', 'Platform_Instagram', 'Platform_LinkedIn', 'Platform_Snapchat', 'Platform_Telegram', 'Platform_Twitter', 'Platform_Whatsapp']].replace({True: 1, False: 0})

# display the resulting DataFrame
df_platform_one_hot_encoded.head()
```

```python
# TATIANA
# ONE-HOT ENCODING

# one-hot encode the Dominant_Emotion column
training_data_cleaned = pd.get_dummies(df_platform_one_hot_encoded, columns=['Dominant_Emotion'], drop_first=False)

# convert True/False to 1/0
training_data_cleaned[['Dominant_Emotion_Anger', 'Dominant_Emotion_Anxiety', 'Dominant_Emotion_Boredom', 'Dominant_Emotion_Happiness', 'Dominant_Emotion_Neutral', 'Dominant_Emotion_Sadness']] = training_data_cleaned[['Dominant_Emotion_Anger', 'Dominant_Emotion_Anxiety', 'Dominant_Emotion_Boredom', 'Dominant_Emotion_Happiness', 'Dominant_Emotion_Neutral', 'Dominant_Emotion_Sadness']].replace({True: 1, False: 0})

# display the resulting DataFrame
print(training_data_cleaned.head())
```

```python
# ORLANDO
# MULTILINEAR REGRESSION

# use the cleaned training data to create a multilinear regression model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# the training data is the training_data_cleaned dataframe

# define the target and features for the training data
# the target is happiness
target = ["Daily_Usage_Time (minutes)"]

# list the features
features_original = ["Age", "Posts_Per_Day", "Likes_Received_Per_Day", "Comments_Received_Per_Day", "Messages_Sent_Per_Day", "Gender_Female", "Gender_Male", "Gender_Non-binary", "Platform_Facebook", "Platform_Instagram",	"Platform_LinkedIn", "Platform_Snapchat", "Platform_Telegram", "Platform_Twitter", "Platform_Whatsapp", "Dominant_Emotion_Anger", "Dominant_Emotion_Anxiety", "Dominant_Emotion_Boredom", "Dominant_Emotion_Happiness", "Dominant_Emotion_Neutral", "Dominant_Emotion_Sadness"]

features_updated = ["Age", "Activity_Score", "Gender_Female", "Gender_Male", "Gender_Non-binary", "Platform_Facebook", "Platform_Instagram",	"Platform_LinkedIn", "Platform_Snapchat", "Platform_Telegram", "Platform_Twitter", "Platform_Whatsapp", "Dominant_Emotion_Anger", "Dominant_Emotion_Anxiety", "Dominant_Emotion_Boredom", "Dominant_Emotion_Happiness", "Dominant_Emotion_Neutral", "Dominant_Emotion_Sadness"]

# define the x and y
X = training_data_cleaned[features_updated]
Y = training_data_cleaned[target]

# split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# create and train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# make predictions on the test data
Y_pred = model.predict(X_test)

# evaluate the model with r^2 and mean squared error
r2 = r2_score(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
print(f"R^2: {r2}")
print(f"Mean Squared Error: {mse}")

# calculate the root mean squared error using the mse value
rmse = mse ** 0.5
print(f"Root Mean Squared Error: {rmse}")


print()

# display the model coefficients
coefficients = pd.DataFrame({'Feature': features_updated, 'Coefficient': model.coef_[0]})
print(coefficients)
```

```python
# TATIANA
# DATA VISUALIZATIONS

# Platform vs Daily Usage (Horizontal Bar Chart)

# Create a dictionary to store platform names and their corresponding average usage time
platforms = ["Platform_Facebook", "Platform_Instagram", "Platform_LinkedIn",
             "Platform_Snapchat", "Platform_Telegram", "Platform_Twitter", "Platform_Whatsapp"]

platform_usage = {}

# Loop through each platform column and calculate the average daily usage for users of that platform
for platform in platforms:
    avg_usage = training_data_cleaned.loc[training_data_cleaned[platform] == 1, "Daily_Usage_Time (minutes)"].mean()
    platform_usage[platform.replace("Platform_", "")] = avg_usage  # Remove "Platform_" prefix for clarity

# Convert dictionary to DataFrame for visualization and sort in descending order
platform_usage_df = pd.DataFrame(list(platform_usage.items()), columns=["Platform", "Average Daily Usage Time"])
platform_usage_df = platform_usage_df.sort_values(by="Average Daily Usage Time", ascending=False)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(y="Platform", x="Average Daily Usage Time", data=platform_usage_df, palette="viridis")

# Customize the plot
plt.ylabel("Social Media Platform")
plt.xlabel("Average Daily Usage Time (minutes)")
plt.title("Average Daily Usage Time by Platform (Sorted)")
plt.xticks(rotation=0)  # Keep x-axis labels horizontal

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define dominant emotions
emotions = ["Dominant_Emotion_Anger", "Dominant_Emotion_Anxiety", "Dominant_Emotion_Boredom",
            "Dominant_Emotion_Happiness", "Dominant_Emotion_Neutral", "Dominant_Emotion_Sadness"]

# Define colors for each emotion
emotion_colors = {
    "Anger": "red",
    "Anxiety": "yellow",
    "Boredom": "saddlebrown",
    "Happiness": "green",
    "Neutral": "orange",
    "Sadness": "blue"
}

# Create a dictionary to store average daily usage for each dominant emotion
emotion_usage = []

# Loop through each emotion column and calculate average daily usage time
for emotion in emotions:
    avg_usage = training_data_cleaned.loc[training_data_cleaned[emotion] == 1, "Daily_Usage_Time (minutes)"].mean()
    emotion_usage.append([emotion.replace("Dominant_Emotion_", ""), avg_usage])

# Convert dictionary to DataFrame for visualization
emotion_usage_df = pd.DataFrame(emotion_usage, columns=["Emotion", "Average Daily Usage Time"])

# Drop any rows with NaN values to avoid issues in visualization
emotion_usage_df = emotion_usage_df.dropna()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(
    x="Emotion",
    y="Average Daily Usage Time",
    data=emotion_usage_df,
    edgecolor="black",
    palette=[emotion_colors[emotion] for emotion in emotion_usage_df["Emotion"]]
)

# Customize the plot
plt.xlabel("Dominant Emotion")
plt.ylabel("Average Daily Usage Time (minutes)")
plt.title("Average Daily Usage Time by Dominant Emotion")
plt.xticks(rotation=45)  # Rotate labels for better readability

# Show the plot
plt.show()
```

```python
# TATIANA
# DATA VISUALIIZATIONS

# x-axis: platform (each with 3 bins: 1 for each gender on that specific platform)

# Define specific colors for each gender
gender_colors = {"Male": "lightblue", "Female": "pink", "Non-binary": "purple"}

# Define platforms and genders
platforms = ["Platform_Facebook", "Platform_Instagram", "Platform_LinkedIn", "Platform_Snapchat", "Platform_Telegram", "Platform_Twitter", "Platform_Whatsapp"]
genders = ["Gender_Male", "Gender_Female", "Gender_Non-binary"]

# Create a list to store the data
platform_gender_usage = []

# Loop through platforms and genders to calculate average daily usage
for platform in platforms:
    for gender in genders:
        avg_usage = training_data_cleaned.loc[
            (training_data_cleaned[platform] == 1) & (training_data_cleaned[gender] == 1),
            "Daily_Usage_Time (minutes)"
        ].mean()
        platform_gender_usage.append([platform.replace("Platform_", ""), gender.replace("Gender_", ""), avg_usage])

# Convert to DataFrame for visualization
usage_df = pd.DataFrame(platform_gender_usage, columns=["Platform", "Gender", "Average Daily Usage Time"])

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    x="Platform",
    y="Average Daily Usage Time",
    hue="Gender",
    data=usage_df, edgecolor="black",
    palette=[gender_colors[g] for g in usage_df["Gender"]]
)

# Customize the plot
plt.xlabel("Social Media Platform")
plt.ylabel("Average Daily Usage Time (minutes)")
plt.title("Average Daily Usage Time by Platform and Gender")
plt.legend(title="Gender")

# Show the plot
plt.show()
```
